```python
import os
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Why do parrots talk")
response.content
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str) -> str:
    """Get the weather in loaction"""
    return f"It's rainy in {location}"

model_with_tools = model.bind_tools([get_weather])
response = model_with_tools.invoke("What's the weather in banglore")
print(response)
```

```python
for tool_call in response.tool_calls:
    print(f"tool: {tool_call['name']}")
    print(f"args: {tool_call['args']}")
```

```python
from langchain.tools import tool

@tool
def add(a:int,b:int) -> int:
    """Addtion of two numbers"""
    return f"Addition of {a} and {b} is {a+b}"

model_with_tools = model.bind_tools([add])
response = model_with_tools.invoke("what's the addition os 2 and 3")
print(response)
```

```python
for tool_call in response.tool_calls:
    print(f"tool: {tool_call["name"]}")
    print(f"args: {tool_call["args"]}")
```

### Tool Execution loops

```python
messages = [{"role":"user","content":"what is the addition of 5 and 6"}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)
```

```python
for tool_call in ai_msg.tool_calls:
    tool_result = add.invoke(tool_call)
    messages.append(tool_result)
```

```python
final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
