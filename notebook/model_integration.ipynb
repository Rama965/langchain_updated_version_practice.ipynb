```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Hey Whats going on!")
response.content
```

### Streaming and batching

```python
model.stream("Why do parrots have colorful features")
```

```python
for chunk in model.stream("Why do parrots have colorful features?"):
    print(chunk.text,end="",flush=True)
```

```python
responses = model.batch([
    "Why do parrots have colorful features",
    "How do airplanes fly",
    "What is quantum computing"
])
for response in responses:
    print(response)
```

```python
model.batch([
    "Why do parrots have colorful features",
    "How do airplanes fly",
    "What is quantum computing",
    "Why do we need to study",
    "What is the job market now?",
    "Who is the richest person in india"],
    config={
        'max_concurrency':5
    }
)
```
